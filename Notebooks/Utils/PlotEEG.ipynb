```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.decomposition import FastICA

directory = "../"

recording_file = directory + "subj-1_ses-S001_task-Diego_Psalm_run-001_20240408_174320_eeg_c2f03392-0d3d-494a-bbf9-b07483aa825a-raw.csv"
data = pd.read_csv(recording_file)

# Elimina la primera columna
data = data.drop(data.columns[0], axis=1)

electrode_columns = data.columns[1:17]

electrodes_data = data[electrode_columns]

display(electrodes_data)

# Definir un filtro, por ejemplo, un filtro pasa baja Butterworth
order = 6
fs = 2000  # Frecuencia de muestreo (Hz)
fc = 50  # Frecuencia de corte (Hz)
b, a = signal.butter(order, fc / (fs / 2), 'low')

electrodes_data_filtered = electrodes_data

for column in electrode_columns:
    electrodes_data_filtered[column] = signal.filtfilt(b, a, electrodes_data[column])

electrodes_data_filtered = pd.DataFrame(electrodes_data_filtered)

display(electrodes_data_filtered)

# Aplicar FastICA a las columnas del EEG
ica = FastICA(n_components=16, random_state=42)

electrodes_data_filtered_ICA = ica.fit_transform(electrodes_data_filtered)

electrodes_data_filtered_ICA = pd.DataFrame(electrodes_data_filtered_ICA)

display(electrodes_data_filtered_ICA)

plt.figure(figsize=(100, 100))
#plt.plot(data[columns_to_filter], color='blue', label='SeÃ±al EEG Original')
plt.plot(electrodes_data_filtered_ICA , color='red', label='SeÃ±al EEG')
plt.title('SeÃ±al EEG')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
